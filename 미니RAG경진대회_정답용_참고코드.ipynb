```python
import re  # 채점 결과 문자열에서 점수만 추출하고, BM25 토큰화에 사용합니다.
from collections import defaultdict  # RRF 점수를 누적할 때 사용합니다.
from pathlib import Path  # 경로를 OS와 무관하게 다룹니다.

from dotenv import load_dotenv  # .env 파일의 환경 변수를 불러옵니다.
from langchain_community.document_loaders import PyPDFLoader  # PDF를 페이지별 Document로 읽습니다.
from langchain_chroma import Chroma  # ChromaDB를 벡터 저장소로 사용합니다.
from langchain_core.output_parsers import StrOutputParser  # LLM 응답에서 문자열만 꺼냅니다.
from langchain_core.prompts import ChatPromptTemplate  # 질문과 문맥을 넣을 프롬프트 틀입니다.
from langchain_openai import ChatOpenAI, OpenAIEmbeddings  # OpenAI LLM과 임베딩 모델을 사용합니다.
from langchain_text_splitters import RecursiveCharacterTextSplitter  # 긴 문서를 검색용 청크로 나눕니다.
from rank_bm25 import BM25Okapi  # 키워드 기반 BM25 검색을 사용합니다.

from grading_utils import evaluation_items  # 질문, 정답, 출처 페이지를 불러옵니다.
from grading_utils import grade_predictions  # 예측 답변을 자동 채점합니다.

load_dotenv(".env", override=True)  # .env 값을 현재 세션에 다시 반영합니다.
```

```python
PDF_PATH = "data/금융투자협회_투자길라잡이_2018.pdf"  # 과제에 사용할 원본 PDF입니다.
DB_PATH = Path("./chroma_index_kofia_guide_solution_prompt_v3")  # 정답 예시용 ChromaDB 저장 경로입니다.
COLLECTION_NAME = "kofia_guide_solution_prompt_v3"  # 정답 예시용 ChromaDB 컬렉션 이름입니다.

CHUNK_SIZE = 500  # 검색용 청크 길이입니다.
CHUNK_OVERLAP = 120  # 청크 사이에 겹치게 둘 길이입니다.
DENSE_TOP_K = 14  # dense 검색에서 가져올 청크 수입니다.
BM25_TOP_K = 14  # BM25 검색에서 가져올 청크 수입니다.
FINAL_CHUNK_K = 16  # RRF 결합 후 사용할 청크 수입니다.
PARENT_PAGE_K = 8  # 검색 청크가 속한 원문 페이지 수입니다.

GENERATOR_MODEL_NAME = "gpt-5-mini"  # 답변 생성에 사용할 모델입니다.
GRADER_MODEL_NAME = "gpt-5-mini"  # 자동 채점 모델입니다.

embedding_model = OpenAIEmbeddings(model="text-embedding-3-small")  # 문서를 벡터로 바꿀 임베딩 모델입니다.
generator_llm = ChatOpenAI(
    model=GENERATOR_MODEL_NAME,
    temperature=0,
    seed=42,
    verbosity="low",
    reasoning_effort="high",
)  # 근거 선택과 최종 표현 선택을 강화합니다.
```

```python
pages = PyPDFLoader(PDF_PATH).load()  # PDF를 페이지 단위로 읽습니다.

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=CHUNK_SIZE,
    chunk_overlap=CHUNK_OVERLAP,
    add_start_index=True,
    separators=["\n\n", "\n", " ", ""],
)
split_docs = text_splitter.split_documents(pages)  # 검색용 청크를 만듭니다.
bm25 = BM25Okapi([re.findall(r"[가-힣A-Za-z0-9§%.]+", doc.page_content.lower()) for doc in split_docs])

if DB_PATH.exists():  # 이미 만든 벡터 DB가 있으면 재사용합니다.
    vectorstore = Chroma(
        collection_name=COLLECTION_NAME,
        embedding_function=embedding_model,
        persist_directory=str(DB_PATH),
    )
else:  # 없으면 PDF 청크로 새 벡터 DB를 만듭니다.
    vectorstore = Chroma.from_documents(
        documents=split_docs,
        embedding=embedding_model,
        collection_name=COLLECTION_NAME,
        persist_directory=str(DB_PATH),
    )
```

```python
def tokenize(text):  # BM25에서 사용할 간단한 토큰화 함수입니다.
    return re.findall(r"[가-힣A-Za-z0-9§%.]+", text.lower())


def doc_key(doc):  # RRF 결합 시 같은 청크를 중복 계산하지 않도록 식별값을 만듭니다.
    page = doc.metadata.get("page")
    start_index = doc.metadata.get("start_index")
    return page, start_index, doc.page_content[:80]


def bm25_search(query, k=BM25_TOP_K):  # 키워드가 잘 맞는 청크를 찾습니다.
    scores = bm25.get_scores(tokenize(query))
    ranked = sorted(enumerate(scores), key=lambda item: item[1], reverse=True)
    return [split_docs[idx] for idx, score in ranked[:k] if score > 0]


def rrf_combine(result_lists, k=60):  # dense 검색과 BM25 검색 결과를 순위 기반으로 결합합니다.
    scores = defaultdict(float)
    docs_by_key = {}
    for docs in result_lists:
        for rank, doc in enumerate(docs, start=1):
            key = doc_key(doc)
            docs_by_key[key] = doc
            scores[key] += 1 / (k + rank)
    ranked_keys = sorted(scores, key=scores.get, reverse=True)
    return [docs_by_key[key] for key in ranked_keys]


def retrieve_docs(query):  # 두 검색 방식을 함께 사용해 문맥 후보를 고릅니다.
    dense_docs = vectorstore.similarity_search(query, k=DENSE_TOP_K)
    keyword_docs = bm25_search(query, k=BM25_TOP_K)
    return rrf_combine([dense_docs, keyword_docs])[:FINAL_CHUNK_K]


def format_supporting_context(docs, max_pages=PARENT_PAGE_K):  # 청크와 원문 페이지를 함께 묶어 문맥을 만듭니다.
    selected_pages = []
    blocks = []
    for doc in docs:
        page = doc.metadata.get("page")
        if isinstance(page, int) and page not in selected_pages:
            selected_pages.append(page)
        page_label = page + 1 if isinstance(page, int) else "?"
        blocks.append(f"[chunk p.{page_label}] {doc.page_content}")

    for page in selected_pages[:max_pages]:
        blocks.append(f"[page p.{page + 1}]\n{pages[page].page_content}")

    return "\n\n".join(blocks)[:20000]


def extract_scores(grading_output: str):  # 채점 결과 문자열에서 문항별 점수만 뽑아냅니다.
    score_pattern = re.compile(r"\d+번:\s*([01](?:\.5)?)점")
    scores = [float(score) for score in score_pattern.findall(grading_output)]
    return sum(scores), scores


evidence_prompt = ChatPromptTemplate.from_template(
    """
QUESTION에 직접 답하는 데 필요한 근거만 CONTEXT에서 고른다.
질문 핵심어와 같은 문단의 근거를 우선하고, 서로 다른 후보가 있으면 둘 다 남긴다.
명칭 뒤에 세부 목록이 이어지는 구조라면 명칭이 들어간 문장도 반드시 포함한다.
괄호 속 약어, 조사, 기간 단위, 서비스명, 경로 기호처럼 답 표현을 결정하는 단서는 생략하지 않는다.
최대 5줄까지만 출력한다.

QUESTION:
{question}

CONTEXT:
{context}

EVIDENCE:
"""
)

candidate_prompt = ChatPromptTemplate.from_template(
    """
QUESTION과 EVIDENCE만 보고 답변 후보 3개를 만든다. 각 후보는 한 줄이어야 한다.
후보에는 EVIDENCE에 없는 사실을 넣지 않는다.

A는 원문 표현을 최대한 그대로 복사한다.
B는 질문이 요구한 항목만 가장 짧게 쓴다.
C는 아래 양식 가이드를 반영해 자동채점용 한 줄 답변으로 쓴다.

양식 가이드:
- 명칭/행위/제도/시스템 이름을 묻는 경우, 세부 설명 목록보다 대표 명칭만 쓴다.
- 시스템·서비스·제도명은 원문의 띄어쓰기를 그대로 보존하고 붙여 쓰지 않는다.
- 대표 명칭 뒤에 콜론과 설명이 붙으면 콜론 앞 명칭만 쓴다.
- 경로는 기관명보다 URL을 앞세우고, 뒤따르는 메뉴를 끝까지 포함한다.
- 인물 라벨과 금액을 묻는 경우 "A는 값, D는 값"처럼 라벨과 조사를 붙인다.
- 처벌 수준과 조항 기호는 "형량, §번호"처럼 쉼표로 구분한다.
- 두 번호 항목을 묻는 경우 첫 항목과 둘째 항목을 " / "로 구분하고, 항목 제목이나 공통 주제어는 반복하지 않는다.
- 긴 구나 문장형 항목 두 개를 묻는 경우 두 항목을 쉼표로 구분하고, 마지막 항목 앞에 "과/와"를 붙이지 않는다.
- 권리 두 개가 하나의 명사구로 답이 되는 경우에는 쉼표보다 "과/와" 같은 자연스러운 연결어를 우선 후보로 만든다.
- 두 가지를 묻는 질문에서는 정확히 두 항목만 만들고, "또는", "및", "과"로 세 번째 항목을 이어 붙이지 않는다.
- "A 또는 B, C 등" 구조에서 두 가지 예시만 요구하면 A와 B를 우선 후보로 만들고 C 이하는 제외한다.
- "기존의 무엇"처럼 대상을 묻는 경우 "기존의"나 뒤의 결과물 표현을 빼고 대상 명사만 후보로 만든다.
- 기간 두 개는 수치 뒤에 역할 단어가 오게 하여 "기간+역할, 기간+역할" 구조로 만든다.
- 비용 배분과 월 비용을 함께 묻는 경우 기준어와 "상당" 같은 수식어를 보존한다.
- 예시 목록은 상위 기준이나 선순위 항목을 빼고, 예시로 제시된 항목만 만든다.
- 사이트 바로 뒤 괄호 안에 함께 제공된 서비스가 있으면 그 서비스만 후보로 만들고, 뒤 문장의 별도 서비스는 섞지 않는다.
- 따옴표, 괄호, 불필요한 번호 표시는 제거한 후보도 만든다.

표현 예시:
- "어떠한 사항에 대해 직원에게 일임을 하는지"는 "직원에게 일임을 하는 사항"으로 쓰고, "일임하는"으로 축약하지 않는다.
- "직원에게 일임을 하는 사항"과 "직원이 해서는 안되는 사항"처럼 긴 항목 두 개는 쉼표로 구분한다.
- 두 권리가 "전환권 행사시점"과 "매수청구권"이면 "전환권 행사시점과 매수청구권"처럼 한 구로 쓴다.
- 정보 제공 서비스는 "A와 B 정보를 전자우편 등으로 제공하며, 제공을 원하지 않을 경우 별도 신청을 통해 제공받지 않을 수 있습니다" 형식으로 쓴다.
- "휴대폰 문자서비스(SMS)"와 "일회용비밀번호(OTP)"처럼 괄호 약어가 있으면 반드시 보존한다.
- "잠금기능" 문장은 "잠금기능을 설정하고 잠금비밀번호는 수시로 변경하기"처럼 조사와 어미를 자연스럽게 두고, 항목 제목의 "모바일폰"은 반복하지 않는다.
- 고객 과실 두 가지를 묻는 경우 "장기간 거래내역 확인 소홀, 계좌번호·비밀번호 관리 소홀"처럼 두 항목만 쉼표로 쓴다.
- 시스템명은 "개인정보 노출자 사고예방 시스템"처럼 원문 띄어쓰기를 살린다.
- 공고 기간과 소요 기간은 "2개월간 공고, 3개월 정도 소요"처럼 수치 다음에 역할을 붙인다.
- 사이트와 서비스는 "URL, 서비스명" 형식으로 쓰고, 같은 문장 괄호 안 서비스만 고른다. 예: "www.msafer.or.kr, 명의도용 사전차단 서비스"
- 배분 비율과 월 가입료는 "가입료의 50%, 월 80만원 상당"처럼 기준어와 상당 표현을 남긴다.

QUESTION:
{question}

EVIDENCE:
{evidence}

출력 형식:
A: ...
B: ...
C: ...
"""
)

select_prompt = ChatPromptTemplate.from_template(
    """
QUESTION, EVIDENCE, 후보 A/B/C를 보고 최종 답변 한 줄만 출력한다.
후보 라벨은 출력하지 않는다. 후보가 모두 부족하면 EVIDENCE에 근거해 한 줄로 직접 작성한다.

QUESTION:
{question}

EVIDENCE:
{evidence}

CANDIDATES:
{candidates}

선택 기준:
- EVIDENCE에 직접 있는 내용만 사용한다.
- 정답은 짧아야 하지만, 질문이 요구한 항목은 모두 있어야 한다.
- 질문이 요구한 개수보다 많은 항목을 넣지 않는다.
- 후보 중 요구 개수에 정확히 맞는 후보가 있으면 그 후보를 우선한다.
- 명칭/행위/제도/시스템 이름을 묻는 질문은 세부 설명이 아니라 대표 명칭만 고른다.
- 시스템·서비스·제도명은 원문의 띄어쓰기를 그대로 보존하고 붙여 쓰지 않는다.
- 후보에 "대표 명칭: 설명"이 있으면 반드시 콜론 앞 대표 명칭만 출력하고, 뒤의 설명 목록은 버린다.
- 경로는 기관명 없이 URL부터 시작하고 뒤따르는 메뉴 전체를 포함한다.
- 인물 라벨과 금액은 "A는 값, D는 값"처럼 쓴다.
- 조항 기호가 있으면 형량 뒤에 쉼표를 두고 §기호를 보존한다.
- 두 번호 항목은 "첫 항목 / 둘째 항목" 형식을 우선하고, 항목 제목이나 공통 주제어는 반복하지 않는다.
- 긴 구나 문장형 항목 두 개는 쉼표로 구분하고, 마지막 항목 앞에 "과/와"를 붙이지 않는다.
- 원문에서 권리 두 개가 하나의 명사구로 함께 제시되면 쉼표보다 "과/와"로 이어진 후보를 우선한다.
- 두 가지를 묻는 질문에서는 정확히 두 항목만 남기고, "또는", "및", "과"로 세 번째 항목을 이어 붙인 후보는 버린다.
- "A 또는 B, C 등" 구조에서 두 가지 예시만 요구하면 A와 B를 남기고 C 이하는 제외한다.
- "기존의 무엇"처럼 대상을 묻는 경우 "기존의"나 뒤의 결과물 표현을 붙이지 않고 대상 명사만 출력한다.
- 기간은 "2개월간 공고, 3개월 정도 소요"처럼 수치 뒤에 역할 단어가 오게 쓴다.
- 배분 비율과 월 비용은 "가입료의 50%, 월 80만원 상당"처럼 기준어와 상당 표현을 보존한다.
- 예시 목록은 "이후에는" 뒤에 이어지는 예시 항목을 우선하고, 그 앞의 기본 기준은 제외한다.
- 사이트 바로 뒤 괄호 안에 함께 제공된 서비스와, 뒤 문장의 별도 서비스를 섞지 않는다.
- 후보가 "URL, 서비스명과 다른 서비스명"처럼 두 서비스를 함께 넣으면 더 짧은 "URL, 서비스명" 후보를 고른다.
- 정보 제공 서비스의 답은 "A와 B 정보를 전자우편 등으로 제공하며, 제공을 원하지 않을 경우 별도 신청을 통해 제공받지 않을 수 있습니다" 형식을 우선한다.
- 괄호 약어가 원문에 있으면 생략하지 않는다.
- 항목 제목에서 이미 주어진 공통어는 답에 반복하지 않는다. 예를 들어 잠금기능 항목 앞에 공통어가 붙으면 공통어를 제거한다.
- "일임을 하는 사항"처럼 원문 조사까지 답 형태를 결정하는 표현은 축약하지 않는다.
- 계산식, 기준 금액, 주변 설명, 따옴표, 불필요한 번호 표시는 제거한다.
- 같은 명칭 후보가 따옴표 있는 형태와 없는 형태로 모두 있으면 따옴표 없는 형태를 고른다.
- 쉼표와 슬래시는 문항의 여러 요소를 구분하는 데만 사용한다.
- 설명, 불릿, 번호, 마침표 없이 한 줄만 출력한다.

FINAL ANSWER:
"""
)


evidence_chain = evidence_prompt | generator_llm | StrOutputParser()
candidate_chain = candidate_prompt | generator_llm | StrOutputParser()
select_chain = select_prompt | generator_llm | StrOutputParser()
```

```python
assert len(evaluation_items) == 20, "평가 문항 수가 20개가 아닙니다."

questions = [item["question"] for item in evaluation_items]
predicted_answers = []  # 20개 질문에 대한 모델 예측 답을 모아둡니다.

for idx, item in enumerate(evaluation_items, start=1):
    question = item["question"]
    context_docs = retrieve_docs(question)
    supporting_context = format_supporting_context(context_docs)

    evidence = evidence_chain.invoke({"question": question, "context": supporting_context}).strip()
    candidates = candidate_chain.invoke({"question": question, "evidence": evidence}).strip()
    answer = select_chain.invoke({"question": question, "evidence": evidence, "candidates": candidates}).strip()

    predicted_answers.append(answer)

    print(f"\n질문 {idx}: {question}")
    print("  예측 답변:", answer)
    print("  정답:", item["answer"])

print("\n✅ 자동 채점 결과")
grading_result = grade_predictions(questions, predicted_answers, model_name=GRADER_MODEL_NAME)
print(grading_result)

total_score, per_item_scores = extract_scores(grading_result)
total_questions = len(evaluation_items)
correct_count = sum(1 for score in per_item_scores if score == 1)
partial_count = sum(1 for score in per_item_scores if score == 0.5)
print(f"\n정답: {correct_count}문항 / {total_questions}문항")
if partial_count:
    print(f"부분 정답: {partial_count}문항")
print(f"총점: {total_score:g}점 / {total_questions}점")
```
