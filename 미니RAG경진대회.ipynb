```python
import re  # 채점 결과 문자열에서 점수만 추출할 때 사용합니다.
from pathlib import Path  # 경로를 OS와 무관하게 다룹니다.

from dotenv import load_dotenv  # .env 파일의 환경 변수를 불러옵니다.
from langchain_community.document_loaders import PyPDFLoader  # PDF를 페이지별 Document로 읽습니다.
from langchain_chroma import Chroma  # ChromaDB를 벡터 저장소로 사용합니다.
from langchain_core.output_parsers import StrOutputParser  # LLM 응답에서 문자열만 꺼냅니다.
from langchain_core.prompts import ChatPromptTemplate  # 질문과 문맥을 넣을 프롬프트 틀입니다.
from langchain_openai import ChatOpenAI, OpenAIEmbeddings  # OpenAI LLM과 임베딩 모델을 사용합니다.
from langchain_text_splitters import RecursiveCharacterTextSplitter  # 긴 문서를 검색용 청크로 나눕니다.

from langfuse.langchain import CallbackHandler  # LangChain 실행 기록을 Langfuse trace로 보냅니다.

from grading_utils import evaluation_items  # 질문, 정답, 출처 페이지를 불러옵니다.
from grading_utils import grade_predictions  # 예측 답변을 자동 채점합니다.

load_dotenv(".env", override=True)  # .env 값을 현재 세션에 다시 반영합니다.
```

```python
PDF_PATH = "data/금융투자협회_투자길라잡이_2018.pdf"  # 과제에 사용할 원본 PDF입니다.
DB_PATH = Path("./chroma_index_kofia_guide")  # 과제용 ChromaDB 저장 경로입니다.
COLLECTION_NAME = "kofia_guide_assignment"  # 과제용 ChromaDB 컬렉션 이름입니다.

CHUNK_SIZE = 500  # 문서를 한 번에 자를 길이입니다.
CHUNK_OVERLAP = 100  # 청크 사이에 겹치게 둘 길이입니다.
TOP_K = 2  # 질문마다 상위 몇 개 청크를 가져올지 정합니다.
GENERATOR_MODEL_NAME = "gpt-5-mini"  # 답변 생성에 사용할 모델입니다.
GRADER_MODEL_NAME = "gpt-5-mini"  # 자동 채점에 사용할 모델입니다.
embedding_model = OpenAIEmbeddings(model="text-embedding-3-small")  # 문서를 벡터로 바꿀 임베딩 모델입니다.
```

```python
if DB_PATH.exists():  # 이미 만든 벡터 DB가 있으면 재사용합니다.
    vectorstore = Chroma(
        collection_name=COLLECTION_NAME,
        embedding_function=embedding_model,  # 저장할 때 쓴 임베딩 모델과 같은 모델을 넘깁니다.
        persist_directory=str(DB_PATH),
    )
else:  # 처음 실행이라면 PDF를 읽어서 새 벡터 DB를 만듭니다.
    loader = PyPDFLoader(PDF_PATH)  # PDF를 페이지 단위로 읽습니다.
    docs = loader.load()  # 전체 페이지를 Document 리스트로 변환합니다.

    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=CHUNK_SIZE,  # 청크 길이를 고정합니다.
        chunk_overlap=CHUNK_OVERLAP,  # 청크가 약간 겹치게 잘라 문맥 손실을 줄입니다.
        separators=["\n\n", "\n", " ", ""],  # 문단, 줄바꿈, 공백 순서로 우선 분할합니다.
    )
    split_docs = text_splitter.split_documents(docs)  # 검색용 청크 리스트를 만듭니다.

    vectorstore = Chroma.from_documents(
        documents=split_docs,
        embedding=embedding_model,  # 청크를 임베딩해 ChromaDB에 저장합니다.
        persist_directory=str(DB_PATH),
        collection_name=COLLECTION_NAME,
    )
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k": TOP_K})  # 질문마다 상위 TOP_K개 청크를 찾습니다.

def format_docs(docs):  # 검색된 청크를 답변 생성용 문맥 문자열로 합칩니다.
    lines = []  # 페이지 번호를 붙인 청크 문자열을 모아둘 리스트입니다.
    for doc in docs:
        page = doc.metadata.get("page")  # 원문 페이지 번호를 가져옵니다.
        prefix = f"[p.{page + 1}] " if page is not None else ""  # 출처 확인이 쉽도록 페이지 번호를 붙입니다.
        lines.append(prefix + doc.page_content)  # 페이지 정보와 본문을 함께 저장합니다.
    return "\n\n".join(lines)  # 청크 사이를 빈 줄로 띄워 읽기 쉬운 문맥으로 만듭니다.

def extract_scores(grading_output: str):  # 채점 결과 문자열에서 문항별 점수만 뽑아냅니다.
    score_pattern = re.compile(r"\d+번:\s*([01](?:\.5)?)점")
    scores = [float(score) for score in score_pattern.findall(grading_output)]
    return sum(scores), scores

langfuse_handler = CallbackHandler()  # 실행 기록을 Langfuse에 남길 콜백입니다.
trace_config = {"callbacks": [langfuse_handler]}

generator_llm = ChatOpenAI(model=GENERATOR_MODEL_NAME)  # 답변 생성에 사용할 모델입니다.
prompt = ChatPromptTemplate.from_template('''
다음은 문서에서 검색된 정보입니다.
{context}

규칙:
1) 질문에 대해 위 문맥에 근거해서만 답하세요.
2) 답은 정답 문자열만 그대로 출력하세요.
3) 조사, 설명, 불릿, 번호, 줄바꿈, 접두어, 접미어를 붙이지 마세요.
4) 여러 요소가 있는 답은 쉼표, 슬래시, 순서까지 문맥의 답 형태를 그대로 따르세요.
5) 문서에 없으면 반드시 정확히 "없는 정보"라고만 답하세요.

질문: {question}
''')  # 질문과 검색 문맥을 함께 넣는 프롬프트입니다.

parser = StrOutputParser()  # 최종 응답에서 문자열만 꺼냅니다.
answer_chain = prompt | generator_llm | parser  # 문맥과 질문을 넣어 답을 만드는 체인입니다.
```

```python
assert len(evaluation_items) == 20, "평가 문항 수가 20개가 아닙니다."

questions = [item["question"] for item in evaluation_items]
predicted_answers = []  # 20개 질문에 대한 모델 예측 답을 모아둡니다.

for idx, item in enumerate(evaluation_items, start=1):
    question = item["question"]
    retrieved_docs = retriever.invoke(question, config=trace_config)[:TOP_K]

    context_text = format_docs(retrieved_docs)
    chain_input = {"question": question, "context": context_text}
    answer = answer_chain.invoke(chain_input, config=trace_config).strip()

    predicted_answers.append(answer)

    print(f"\n질문 {idx}: {question}")
    print("  예측 답변:", answer)
    print("  정답:", item["answer"])

print("\n✅ 자동 채점 결과")
grading_result = grade_predictions(questions, predicted_answers, model_name=GRADER_MODEL_NAME)
print(grading_result)

total_score, per_item_scores = extract_scores(grading_result)
total_questions = len(evaluation_items)
correct_count = sum(1 for score in per_item_scores if score == 1)
partial_count = sum(1 for score in per_item_scores if score == 0.5)
print(f"\n정답: {correct_count}문항 / {total_questions}문항")
if partial_count:
    print(f"부분 정답: {partial_count}문항")
print(f"총점: {total_score:g}점 / {total_questions}점")
```
